```python
# read csv file from VNstock folder
from dataloader import *

collector = OurDatasetCollector()
df = collector.ReadFromFolder("VNstock")
df = df[(df['date'] >= TEST_START_DATE) & (df['date'] <= TEST_END_DATE)]
df = df.sort_values(by='date', ascending=True)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.figure(figsize=(10, 5))

class UpdateDrawing:
    def __init__(self, df):
        self.stocks = [0, 0, 0, 0, 0]
        self.df = df
    def update(self, frame):
        plt.clf()
        df = self.df
        if frame == 0:
            return
        else:
            print(f'{frame}/{len(df)//5}')

        plt.subplot(5, 1, 1)    
        ticker_data = df[df['tic'] == 'ACB']
        date = ticker_data['date'][:frame]
        plt.plot(ticker_data['date'][:frame], ticker_data['close'][:frame], label='ACB', color='red')
        plt.legend()
        plt.xticks([])
        plt.tight_layout()

        plt.subplot(5, 1, 2)    
        ticker_data = df[df['tic'] == 'BID']
        date = ticker_data['date'][:frame]
        plt.plot(ticker_data['date'][:frame], ticker_data['close'][:frame], label='BID', color='green')
        plt.legend()
        plt.xticks([])
        plt.tight_layout()

        plt.subplot(5, 1, 3)    
        ticker_data = df[df['tic'] == 'BVH']
        date = ticker_data['date'][:frame]
        plt.plot(ticker_data['date'][:frame], ticker_data['close'][:frame], label='BVH', color='blue')
        plt.legend()
        plt.xticks([])
        plt.tight_layout()

        plt.subplot(5, 1, 4)    
        ticker_data = df[df['tic'] == 'FPT']
        date = ticker_data['date'][:frame]
        plt.plot(ticker_data['date'][:frame], ticker_data['close'][:frame], label='FPT', color='yellow')
        plt.legend()
        plt.xticks([])
        plt.tight_layout()


        plt.subplot(5, 1, 5)    
        ticker_data = df[df['tic'] == 'VIC']
        date = ticker_data['date'][:frame]
        plt.plot(ticker_data['date'][:frame], ticker_data['close'][:frame], label='VIC', color='black')
        plt.legend()
        plt.xticks(rotation=0)
        num_dates = len(date)
        n = max(num_dates // 5, 1)
        plt.xticks(date.values[::n])
        plt.tight_layout()
        

myupdate = UpdateDrawing(df)
ani = FuncAnimation(plt.gcf(), myupdate.update, frames=len(df)//5, interval=200)  # Each frame runs for 0.05 seconds
ani.save('stock.gif', writer='ffmpeg')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Assuming your DataFrame is named df
action = pd.read_csv("merged_actions.csv")
action['date'] = pd.to_datetime(action['date'])

account_value = pd.read_csv("merged_account_value.csv")

plt.figure(figsize=(10, 5))

class UpdateDrawing:
    def __init__(self):
        self.stocks = [0, 0, 0, 0, 0]

    def update(self, frame):
        stocks = [0, 0, 0, 0, 0]
        plt.clf()
        if frame == 0:
            return
        
        A = action.at[frame, 'ACB']
        B = action.at[frame, 'BID']
        C = action.at[frame, 'BVH']
        D = action.at[frame, 'FPT']
        E = action.at[frame, 'VIC']
        date = action.at[frame, 'date'].strftime('%Y-%m-%d')  # Format the date to show only year-month-day
        value = account_value.at[frame, 'account_value']

        if A > 0:
            self.stocks[0] += A
            BuyA = A
            sellA = 0
        else:
            BuyA = 0
            sellA = -A
            self.stocks[0] -= A

        if B > 0:
            self.stocks[1] += B
            BuyB = B
            sellB = 0
        else:
            BuyB = 0
            sellB = -B
            self.stocks[1] -= B

        if C > 0:
            self.stocks[2] += C
            BuyC = C
            sellC = 0
        else:
            BuyC = 0
            sellC = -C
            self.stocks[2] -= C

        if D > 0:
            self.stocks[3] += D
            BuyD = D
            sellD = 0
        else:
            BuyD = 0
            sellD = -D
            self.stocks[3] -= D

        if E > 0:
            self.stocks[4] += E
            BuyE = E
            sellE = 0
        else:
            BuyE = 0
            sellE = -E
            self.stocks[4] -= E

        plt.subplot2grid((2, 3), (0, 0))
        plt.text(0.5, 0.5, f"Date:\n{date}\nAccount value:\n{round(value,2)}", ha='right', va='center', fontsize=18, fontname='Times New Roman')
        plt.axis('off')

        plt.subplot2grid((2, 3), (0, 1), colspan=2)
        plt.plot(account_value['date'][:frame], account_value['account_value'][:frame], color='red')
        plt.xticks([])
        # plt.axis('off')

        plt.subplot2grid((2, 3), (1, 0))
        plt.text(0.5, 0.5, f"Buy stocks\nACB: {BuyA}\nBID: {BuyB}\nBVH: {BuyC}\nFPT: {BuyD}\nVIC: {BuyE}",
                 ha='center', va='center', fontsize=16, fontname='Times New Roman')
        plt.axis('off')

        plt.subplot2grid((2, 3), (1, 1))
        plt.text(0.5, 0.5, f"Sell stocks\nACB: {sellA}\nBID: {sellB}\nBVH: {sellC}\nFPT: {sellD}\nVIC: {sellE}",
                 ha='center', va='center', fontsize=16, fontname='Times New Roman')
        plt.axis('off')

        # State 3: Empty state
        plt.subplot2grid((2, 3), (1, 2))
        plt.text(0.5, 0.5, f"Holding\nACB: {self.stocks[0]}\nBID: {self.stocks[1]}\nBVH: {self.stocks[2]}\nFPT: {self.stocks[3]}\nVIC: {self.stocks[4]}",
                 ha='center', va='center', fontsize=16, fontname='Times New Roman')
        plt.axis('off')

myupdate = UpdateDrawing()
ani = FuncAnimation(plt.gcf(), myupdate.update, frames=len(action), interval=200)  # Each frame runs for 0.05 seconds
ani.save('trading.gif', writer='ffmpeg')
```
